```python
class chatbot():
  def __init__(self, model:str='gpt-4', temperature:float=1, top_p:float=1):
    self.model = model
    self.temperature = temperature
    self.top_p = top_p

  def positive_chat(self, content):
    pos_chat = openai.ChatCompletion.create(
      model = self.model,
      temperature = self.temperature,
      top_p = self.top_p,
      messages=[
        {"role": "system", "content": "당신은 긍정적인 리뷰에 답변을 달아주는 유용한 AI 봇입니다. 상대방의 칭찬에 감사와 고마움을 표하는 답변을 작성하세요."},
        {"role": "user", "content": "믿고쓰는 상품! 너무나도 만족합니다. 항상 좋은 제품 제공해주셔서 너무나도 감사합니다. 배송도 빠르고 서비스도 너무 좋아요 ^^"},
        {"role": "assistant", "content": "안녕하세요 고객님, 항상 저희 제품을 사용해주셔서 대단히 감사합니다. 앞으로도 좋은 서비스로 보답할 수 있도록 하겠습니다. 고맙습니다. "},
        {"role": "user", "content": f"다음의 리뷰에 대한 답변을 작성하시오 /n {content}"}
    ]
    )
    return (pos_chat.choices[0].message['content'])

  def negative_chat(self, content):
    neg_chat = openai.ChatCompletion.create(
      model = self.model,
      temperature = self.temperature,
      top_p = self.top_p,
      messages=[
        {"role": "system", "content": "당신은 부정적인 리뷰에 답변을 달아주는 유용한 AI 봇입니다. 고객의 마음을 이해하고 위로하는 답변을 작성하세요."},
        {"role": "user", "content": "예쁘고 심플해서 샀는데. 재질이 깔끄러워요. 살에 자국 다 베이고ㅠㅠ....폭망이에요. 재대로 확인안한 제 잘못이죠;;; 참고로 싱글세트 2. 퀸세트 1 샀습니다."},
        {"role": "assistant", "content": "안녕하세요 고객님, 먼저 저희 제품을 선택해 주신 것에 대한 감사함을 먼저 표합니다. 싱글세트와 퀸세트를 구매해주셨는데, 의도치않게 불편을 드리게 되어 정말로 죄송합니다. 추후에는 이러한 부분을 보완하여 더욱 좋은 상품을 제공할 수 있도록 노력하겠습니다. 감사합니다."},
        {"role": "user", "content": f"다음의 리뷰에 대한 답변을 작성하시오 /n {content}"}
    ]
    )
    return (neg_chat.choices[0].message['content'])
```

```python
chatbot = chatbot()
chatbot.positive_chat('빠른 배송과 친절한 설명')
```

```python
chatbot = chatbot()
chatbot.negative_chat("피부 트러블 더 많이 올라와서 몸에 사용했는데 몸에도 뾰루지나네요ㅜ")
```
